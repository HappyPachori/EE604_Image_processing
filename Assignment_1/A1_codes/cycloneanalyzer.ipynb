#### Function for drawing line passing through two given points

```python
def drawLine(image,x1,y1,x2,y2):

    m=(y2-y1)/(x2-x1)
    h,w=image.shape[:2]
    
    px=0
    py=-(x1-0)*m+y1
    qx=w
    qy=-(x2-w)*m+y2
    cv2.line(image, (int(px), int(py)), (int(qx), int(qy)), (255, 0, 0), 2)
```

```python
drawLine(image,223,121,217,134)
drawLine(image,539,189,528,196)
```

```python
## code to find the intersecting point of two lines
def findIntersection(x1,y1,x2,y2,x3,y3,x4,y4):
        px= ( (x1*y2-y1*x2)*(x3-x4)-(x1-x2)*(x3*y4-y3*x4) ) / ( (x1-x2)*(y3-y4)-(y1-y2)*(x3-x4) ) 
        py= ( (x1*y2-y1*x2)*(y3-y4)-(y1-y2)*(x3*y4-y3*x4) ) / ( (x1-x2)*(y3-y4)-(y1-y2)*(x3-x4) )
        return [px, py]
```
